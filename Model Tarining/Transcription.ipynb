```python
#!/usr/bin/env python3
# Requires PyAudio and PySpeech.
import os
import time

import speech_recognition as sr

files = [f for f in os.listdir('.') if os.path.isfile(f) and f.endswith(".wav")]
print(files)
numfiles = len(files)

# Record Audio
r = sr.Recognizer()
file_object = open('transcribe_I191706.txt', 'a',encoding="utf-8")
count = 1
for file in files:
    print(file +"   ===>    "+ str(count) + " out of " + str(numfiles))


    with sr.AudioFile(file) as source:
        audio = r.listen(source)

    # Speech recognition using Google Speech Recognition
    try:
        # for testing purposes, we're just using the default API key
        # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
        # instead of `r.recognize_google(audio)`
        result = r.recognize_google(audio, language='ur-PK')
        print("File " + str(count) + " Result: "+ str(result))

        file_object.write(result+"\n")
        count+=1
        time.sleep(10)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        count+=1
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
        count+=1
file_object.close()
```
